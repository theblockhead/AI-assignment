```python
import pandas as pd
import re

def split_into_sentences(text):
    # Split the text into sentences using regex
    sentences = re.split(r'(?<=[.!?])\s+', text)
    return sentences

def join_sentences_to_paragraphs(sentences, sentences_per_paragraph=40):
    """
    Joins a list of sentences into paragraphs, where each paragraph contains
    the specified number of sentences.

    Args:
        sentences (list): List of sentences.
        sentences_per_paragraph (int): Number of sentences per paragraph.

    Returns:
        list: List of paragraphs.
    """
    paragraphs = []
    current_paragraph = ""

    for i, sentence in enumerate(sentences):
        current_paragraph += sentence + " "

        if (i + 1) % sentences_per_paragraph == 0:
            paragraphs.append(current_paragraph.strip())
            current_paragraph = ""

    # Add any remaining sentences as the last paragraph
    if current_paragraph:
        paragraphs.append(current_paragraph.strip())

    return paragraphs

# Example DataFrame with a column named 'text' (replace with your actual DataFrame)
csv_file_path = "raw_text.csv"  # Replace with your actual CSV file path
df = pd.read_csv(csv_file_path)


# Split each element in the 'text' column into sentences
df["sentences"] = df["text"].apply(split_into_sentences)

# Join sentences into paragraphs (40 sentences per paragraph)
df["paragraphs"] = df["sentences"].apply(join_sentences_to_paragraphs, sentences_per_paragraph=40)

new_df = pd.DataFrame({"paragraphs": df["paragraphs"].explode()})

# Save the new DataFrame to a CSV file (replace with your desired filename)
new_df.to_csv("paragraphs.csv", index=False)

print(f"Paragraphs created and saved to paragraphs.csv with {40} sentences per paragraph.")

# Print the modified DataFrame
print(df[["text", "paragraphs"]])
```

```python
import requests

def get_page_views(page_title):
    base_url = "https://wikimedia.org/api/rest_v1/metrics/pageviews/per-article/en.wikipedia/all-access/all-agents/"
    endpoint = f"{page_title}/daily/2015100100/2015103100"  # Example date range

    try:
        response = requests.get(base_url + endpoint)
        data = response.json()
        view_count = data["items"][0]["views"]
        return view_count
    except Exception as e:
        print(f"Error retrieving view count: {e}")
        return 0

# Example usage
page_title = "America"
views = get_page_views(page_title)
print(f"View count for '{page_title}': {views}")
```
